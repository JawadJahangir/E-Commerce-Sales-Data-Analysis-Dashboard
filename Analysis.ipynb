```python
#first import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "Month":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug"],
    "Sales":[12000,15000,18000,16000,21000,25000,27000,30000],
    "Orders":[120,140,160,150,180,200,210,230],
    "Ads":[2000,2500,3000,2800,3200,3500,3800,4000],
}

df = pd.DataFrame(data)
df.to_csv("Dataset.csv")

categories = ["Electronics","Clothing","Home","Beauty"]
category_sales = [45000,30000,20000,15000]

ages = np.random.randint(18,60,200)
```

```python
#Show monthly sales trend.
plt.plot(df["Month"],df["Sales"],color="Yellow",ls="--",lw=2,marker="o")
plt.title("Monthly sales Trend",fontsize=14,color="grey",fontweight="bold")
plt.xlabel("Months",fontsize=12,color="grey",labelpad=10)
plt.ylabel("Sales",fontsize=12,color="grey",labelpad=10)
plt.show()
```

```python
#Show category comparison.
plt.bar(categories,category_sales,color="Yellow")
plt.title("Sales Comparison On Basis of Categories",fontsize="14",color="grey",fontweight="bold")
plt.xlabel("Categories",fontsize="12",color="grey",labelpad=10)
plt.ylabel("Sales",fontsize="12",color="grey",labelpad=10)
plt.show()
```

```python
#Orders by Month
plt.barh(df["Month"],df["Orders"],color="Yellow")
plt.title("Orders bY Month",fontsize=14,color="grey",fontweight="bold")
plt.xlabel("Orders",fontsize=12,color="grey",labelpad=10)
plt.ylabel("Months",fontsize=12,color="grey",labelpad=10)
plt.show()
```

```python
#Category Revenue Share
explode=[0.1,0,0,0]
plt.pie(category_sales,labels=categories,autopct="%1.f%%",textprops={"color":"grey"},explode=explode,shadow=True)
plt.title("Category Revenue Share",fontsize=14,color="grey",fontweight="bold")
plt.show()
```

```python
#Revenue vs Advertising
plt.stackplot(df["Ads"],df["Sales"],labels=["Sales","Ads"],color=["Yellow","red"])
plt.title("Revenue vs Advertising",fontsize=14,color="grey",fontweight="bold")
plt.xlabel("Ads",fontsize=12,color="grey",labelpad=10)
plt.ylabel("Sales",fontsize=12,color="grey",labelpad=10)
plt.grid()
plt.legend()
```

```python
#Ads vs Sales
sizes=[size*2 for size in range(df["Sales"].count())]
color=["red" if sales<20000 else "green" for sales in df["Sales"]]
plt.scatter(df["Ads"],df["Sales"],marker="o",s=100,color=color)
plt.title("Ads vs Sales",fontsize=14,color="grey",fontweight="bold")
plt.xlabel("Ads",fontsize=12,color="grey",labelpad=10)
plt.ylabel("Sales",fontsize=12,color="grey",labelpad=10)
plt.grid()
plt.show()
```

```python
#Customer Age Distribution
plt.hist(ages,color="Yellow",edgecolor="black",bins=20)
plt.title("Customer Age Distribution",fontsize=14,color="grey",fontweight="bold")
plt.xlabel("Age",fontsize=12,color="grey",labelpad=10)
plt.ylabel("Frequency",fontsize=12,color="grey",labelpad=10)
plt.show()
```

```python
#Subplot Dashboard
fig,axs=plt.subplots(2,2,figsize=(10,5))
axs[0,0].plot(df["Month"],df["Sales"],color="Yellow",ls="--",lw=2,marker="o")
axs[0,0].set_title("Monthly sales Trend",fontsize=14,color="grey",fontweight="bold")
axs[0,0].set_xlabel("Months",fontsize=12,color="grey",labelpad=10)
axs[0,0].set_ylabel("Sales",fontsize=12,color="grey",labelpad=10)

axs[0,1].bar(categories,category_sales,color="Yellow")
axs[0,1].set_title("Sales Comparison On Basis of Categories",fontsize=14,color="grey",fontweight="bold")
axs[0,1].set_xlabel("Categories",fontsize=12,color="grey",labelpad=10)
axs[0,1].set_ylabel("Sales",fontsize=12,color="grey",labelpad=10)

axs[1,0].scatter(df["Ads"],df["Sales"],marker="o",s=100,color=color)
axs[1,0].set_title("Ads vs Sales",fontsize=14,color="grey",fontweight="bold")
axs[1,0].set_xlabel("Ads",fontsize=12,color="grey",labelpad=10)
axs[1,0].set_ylabel("Sales",fontsize=12,color="grey",labelpad=10)
axs[1,0].grid()

axs[1,1].hist(ages,color="Yellow",edgecolor="black",bins=20)
axs[1,1].set_title("Customer Age Distribution",fontsize=14,color="grey",fontweight="bold")
axs[1,1].set_xlabel("Age",fontsize=12,color="grey",labelpad=10)
axs[1,1].set_ylabel("Frequency",fontsize=12,color="grey",labelpad=10)

fig.suptitle("Ecommerce Sales Dashboard",fontsize=16,color="Orange",fontweight="bold")
fig.tight_layout()
plt.show()
```
